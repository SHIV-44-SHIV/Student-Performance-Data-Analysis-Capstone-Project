```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:/Users/ADMIN/Downloads/archive/StudentsPerformance.csv")  # Change the path if needed

# Show the first 5 rows
df.head()
```

```python
# Structure of the dataset
df.info()

# Column names
print("\nðŸ“‹ Columns in dataset:")
print(df.columns.tolist())
```

```python
# Check missing values
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# Summary of numeric columns
df.describe()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
```

```python
# Average scores by gender
gender_scores = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean()

# Plot
gender_scores.plot(kind='bar', figsize=(10,6), title="Average Scores by Gender")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# Compare scores of students who completed test preparation vs those who didn't
prep_scores = df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].mean()

# Plot
prep_scores.plot(kind='bar', figsize=(10,6), title="Scores Based on Test Preparation Course")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# Calculate average score
df['average_score'] = (df['math_score'] + df['reading_score'] + df['writing_score']) / 3

# Plot average score by parental education
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='parental_level_of_education', y='average_score')
plt.title('Parental Education vs Student Average Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['math_score', 'reading_score', 'writing_score', 'average_score']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Score Variables")
plt.show()
```

```python
# Save the cleaned and updated dataframe to a new CSV
df.to_csv("cleaned_student_data.csv", index=False)
print("âœ… CSV Exported Successfully!")
```

```python
import pandas as pd
import sqlite3

# Load CSV file
df = pd.read_csv("C:/Users/ADMIN/Downloads/archive/StudentsPerformance.csv")  # âœ… Make sure the file is in the same folder

# Connect to SQLite database (it will create one if not exists)
conn = sqlite3.connect("student_performance.db")

# Load the CSV data into a new table called StudentsPerformance
df.to_sql("StudentsPerformance", conn, if_exists="replace", index=False)

# Confirm the table is created
tables = conn.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
print("âœ… Tables in the database:")
for table in tables:
    print(table[0])

# Done
conn.close()
```

```python
import sqlite3
import pandas as pd

# Connect to the existing SQLite DB
conn = sqlite3.connect("student_performance.db")

# Run SQL query
query = """
SELECT "gender", 
       ROUND(AVG("math score"), 2) AS avg_math, 
       ROUND(AVG("reading score"), 2) AS avg_reading,
       ROUND(AVG("writing score"), 2) AS avg_writing
FROM StudentsPerformance
GROUP BY "gender";
"""


# Fetch the result into a DataFrame
df_result = pd.read_sql_query(query, conn)
print(df_result)

# Optionally export this to a CSV for Power BI
df_result.to_csv("gender_wise_scores.csv", index=False)

# Done
conn.close()
```
